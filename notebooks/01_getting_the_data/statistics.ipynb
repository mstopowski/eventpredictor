# Imports

```python
import os
import sys
import glob
import datetime

import pandas as pd
```

```python
sys.path.append(os.path.abspath('../../'))
sys.path.append(os.path.abspath('../../eventpredictor'))

import eventpredictor.config as paths
from eventpredictor.data.download_data import FootballScraper as fs
```

```python
df = pd.read_csv(os.path.join(paths.STATISTICS, 'statistics.csv'))
```

```python
df.head()
```

```python
cols_to_select = ['id']
```

```python
for col in df.columns:
    if '_all' in col and 'Value' not in col:
        cols_to_select.append(col)
```

```python
df = df.loc[:, cols_to_select]
```

```python
df['home_ball_possession_all'] = df['home_ball_possession_all'].str.split('%').str.get(0).astype('float')
df['away_ball_possession_all'] = df['away_ball_possession_all'].str.split('%').str.get(0).astype('float')
```

```python
df.to_csv(os.path.join(paths.STATISTICS, 'statistics_filtered.csv'), index=False)
```
